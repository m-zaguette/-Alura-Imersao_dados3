```python
import pandas as pd
import numpy as np
import matplotlib as plot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv("./imersaodados3/dados/dados_experimentos/dados_experimentos.csv")
dados.head() #Primeiras 5 linhas dos dados
```

```python
mapa = {"droga":"composto"}
dados.rename(columns = mapa, inplace=True)
```

```python
dados.head()
```

```python
codCompostos = dados["composto"].value_counts().index[0:5]
print(codCompostos)
```

```python
dados.query("composto in @codCompostos")
```

```python
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = "composto", data=dados.query("composto in @codCompostos"))
ax.set_title("Top 5 Compostos")
plt.show()
```
